```python
#Task: Price Range Analysis
# import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\JOB and Internship\internship congify\Dataset .csv')
df.head()
```

```python
df['Price range'].value_counts()
```

```python
#Determine the most common price range among all the restaurants
most_common_price = df['Price range'].value_counts().idxmax()
print("The most common price range among all the restaurants is:", most_common_price)
if(most_common_price == 1):
    print("Most restaurants are budget-friendly.")
elif(most_common_price == 2):
    print("Most restaurants are mid-range.")
elif(most_common_price == 3 or most_common_price == 4):
    print("Most restaurants are premium.")
```

```python
#Calculate the average rating for each price

avg_rating_per_price = df.groupby("Price range")["Aggregate rating"].mean()

# Print results
print("Average Rating for Each Price Range:")
print(avg_rating_per_price)
```

```python
df['Rating color'].value_counts()
```

```python
df['Aggregate rating'].value_counts()
```

```python
#Identify the color that represents the highest average rating among different price ranges
highest_avg_rating = avg_rating_per_price.idxmax()
if(highest_avg_rating == 1):
    print("The color green represents the highest average rating among different price ranges.")
elif(highest_avg_rating == 2):
    print("The color orange represents the highest average rating among different price ranges.")
elif(highest_avg_rating == 3):
    print("The color red represents the highest average rating among different price ranges.")
elif(highest_avg_rating == 4):
    print("The color maroon represents the highest average rating among different price ranges.")
```

```python
#Price vs Rating Scatter Plot 
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price range', y='Aggregate rating', data=df, color='blue')
plt.title('Price Range vs Rating')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10, 6))

# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
